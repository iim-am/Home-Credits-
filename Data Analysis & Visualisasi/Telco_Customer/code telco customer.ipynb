$\textbf{Now how if we want to make a single plot in a lot of axes?}$

This is where $\textbf{subplots}$ does its job. The documentation can be seen [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html)

Create a figure and a set of subplots. This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.

## X-Y Limit

$\textbf{Using xlim(), ylim()}$

In this exercise, you will work with the matplotlib.pyplot interface to quickly set the x- and y-limits of your plots.

You will now create the same figure as in the previous exercise using plt.plot(), this time setting the axis extents using plt.xlim() and plt.ylim(). These commands allow you to either zoom or expand the plot or to set the axis ranges to include important values (such as the origin).

In this exercise, as before, the percentage of women graduates in Computer Science and in the Physical Sciences are held in the variables computer_science and physical_sciences respectively over year.

```python
# Plot the % of degrees awarded to women in Computer Science and the Physical Sciences
plt.plot(year,computer_sciences, color='red') 
plt.plot(year, physical_sciences, color='blue')

# set the limit
plt.xlim(1990,2010)
plt.ylim(0, 50)

# to set x axis range
plt.xticks(range(1990, 2014, 4));
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8,5))

# Plot the % of degrees awarded to women in Computer Science and the Physical Sciences
ax.plot(year,computer_sciences, color='red') 
ax.plot(year, physical_sciences, color='blue')

# Set the x-axis range
ax.set_xlim(1990, 2010)
ax.set_xticks(range(1990, 2014, 4))

# Set the y-axis range
ax.set_ylim(0, 50)

# adjusts subplot params so that the subplot(s) fits in to the figure area
plt.tight_layout()
```

## Legend

$\textbf{Using legend()}$ documentation [here](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html)

Legends are useful for distinguishing between multiple datasets displayed on common axes. The relevant data are created using specific line colors or markers in various plot commands. Using the keyword argument label in the plotting function associates a string to use in a legend.

For example, here, you will plot enrollment of women in the Physical Sciences and in Computer Science over time. You can label each curve by passing a label argument to the plotting call, and request a legend using plt.legend(). Specifying the keyword argument loc determines where the legend will be placed.

```python
# Specify the label 'Computer Science'
plt.plot(year, computer_sciences, color='red', label='Computer Science') 

# Specify the label 'Physical Sciences' 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')

# Add a legend at the lower center
plt.legend(loc='lower center');
```

```python
fig, axs = plt.subplots(2, # number of rows
                     1, # number of columns
                     figsize=(8, 5))

axs[0].plot(year, physical_sciences, color ='blue')
axs[0].plot(year, business, color = 'red')
axs[0].legend(loc = 'lower right')

axs[1].plot(year, physical_sciences, color ='blue')
axs[1].plot(year, computer_sciences, color ='darkorange');
```

## Using Annotate

$\textbf{Using annotate()}$ documentation [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html)

It is often useful to annotate a simple plot to provide context. This makes the plot more readable and can highlight specific aspects of the data. Annotations like text and arrows can be used to emphasize specific observations.

Here, you will once again plot enrollment of women in the Physical Sciences and Computer science over time. The legend is set up as before. Additionally, you will mark the inflection point when enrollment of women in Computer Science reached a peak and started declining using plt.annotate().

To enable an arrow, set arrowprops=dict(facecolor='black'). The arrow will point to the location given by xy and the text will appear at the location given by xytext.

```python
computer_sciences.max()
```

```python
computer_sciences.idxmax()
```

```python
year[computer_sciences.idxmax()]
```

```python
# Plot with legend as before
plt.plot(year, computer_sciences, color='red', label='Computer Science') 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
# plt.legend(loc='lower right')

# Compute the maximum enrollment of women in Computer Science: cs_max
cs_max = computer_sciences.max()

# Calculate the year in which there was maximum enrollment of women in Computer Science: yr_max
yr_max = year[computer_sciences.idxmax()]

# Add a black arrow annotation
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), 
             arrowprops=dict(
                             ec='black', # color arrow
                             arrowstyle='simple', #style of arrow
#                              connectionstyle='angle3', # angle of arrow
                             lw=1 # width of arrow
             )
                            );
```

Let's looping this, let say we want to annotate maximum value on every categories of degrees! How can we do it?

The categories we want to annotate are physcial sciences, computer sciences, & business

```python
fig, ax = plt.subplots(1, 1, figsize = (8,5))
x = year
ys = [physical_sciences, computer_sciences, business]

for y in ys:
    ax.plot(x, y, label=y.name)
    
    cs_max = y.max()
    yr_max = x[y.argmax()]
    ax.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), 
            arrowprops=dict(facecolor='black'))
    ax.legend(loc='lower right')
    
ax.set_xlabel('Year')
ax.set_ylabel('Enrollment (%)')
ax.set_title('Undergraduate enrollment of women')
ax.set_ylim(5, 60) # some minor adjustment
plt.savefig('fig/undergraduate_enrollment_of_women.png');
```

$\textbf{another example!}$

Let say we want to count how many unique customer who churn and not, based on payment method!

```python
## MATPLOTLIB VERSION ##
import numpy as np

dfg2 = df.groupby(['PaymentMethod', 'Churn'])['customerID'].nunique().reset_index(name='unique_customer')
dfg_piv = dfg2.pivot_table(index='PaymentMethod', columns='Churn', values='unique_customer').reset_index()

fig,ax = plt.subplots(1,1,figsize=(10,8))

segmentA = dfg_piv['Yes']
segmentB = dfg_piv['No']

#barplot
bars = dfg_piv['PaymentMethod']
barwidth = 0.4

total_pos = np.arange(len(bars))
segA_pos = [x + barwidth for x in total_pos]
segB_pos = [x + 2*barwidth for x in total_pos]

barlist = plt.bar(segA_pos,segmentA,width=barwidth,label='Churn', color='darkorange')
barlist2 = plt.bar(segB_pos,segmentB,width=barwidth,label='No Churn', color='steelblue')

plt.legend(loc='upper left')
plt.xticks(segA_pos, bars,rotation=0)
plt.yticks();
```

```python
## SEABORN VERSION ##

# we can use countplot since one row represent one unique customer
plt.figure(figsize = (10, 8))
sns.countplot(x='PaymentMethod', hue='Churn', data=df);
```

This function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. The kind parameter selects the underlying axes-level function to use:
